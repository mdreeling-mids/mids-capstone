```python
import sys
import pandas as pd

def print_dataframe_sizes():
    """
    Find all DataFrames in the current notebook environment and print their memory usage in MB.
    """
    print("üìä DataFrame Memory Usage (MB):\n")
    
    for var_name, var_value in globals().items():
        if isinstance(var_value, pd.DataFrame):  # Check if it's a DataFrame
            mem_usage_mb = var_value.memory_usage(deep=True).sum() / (1024 ** 2)  # Convert bytes to MB
            print(f"{var_name}: {mem_usage_mb:.2f} MB")

# Call the function to print sizes
print_dataframe_sizes()
```

# One-hot encode the categorical variables

```python
import gc
# Define categorical columns to encode
categorical_cols = ["ST349Q01JA", "LANGN", "SC177Q01JA", "SC177Q02JA", "SC177Q03JA", "MATHEXC", "SCHLTYPE"]

chunk_size = 100_000  # Adjust based on available RAM
chunk_list = []  # Store processed chunks

for i in range(0, len(student_dataset), chunk_size):
    chunk = student_dataset.iloc[i:i+chunk_size]  # Get chunk
    chunk_encoded = pd.get_dummies(chunk, columns=categorical_cols, dtype='uint8')  # One-hot encode
    chunk_list.append(chunk_encoded)  # Store in list
    del chunk, chunk_encoded  # Free memory
    gc.collect()

# Reassemble the DataFrame
temp3 = pd.concat(chunk_list, ignore_index=True)
```

```python
# Rename columns by removing ".0" at the end
temp3.rename(columns=lambda x: x[:-2] if x.endswith(".0") else x, inplace=True)
```
